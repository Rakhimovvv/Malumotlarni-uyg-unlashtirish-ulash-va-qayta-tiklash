```python
a = pd.DataFrame({"Transport" :["lacetty" , "malibu" , "cobalt"] ,
                  "narxi" :[400 , 500 , 600]})
a
```

```python
b = pd.DataFrame({"Transport" :["lacetty" , "malibu" , "nexia"] ,
                  "narxi" :[400 , 500 , 300]})
b
```

```python
merged = pd.merge(a,b)
merged
```

```python
merged_outer = pd.merge(a,b, how="outer")
merged_outer
```

```python
merged_left = pd.merge(a,b, how="left")
merged_left
```

```python
merged_right = pd.merge(a,b, how="right")
merged_right
```

```python
a = a.set_index("Transport")
a
```

```python
merged1 = pd.merge(a,b , left_index = True , right_on="Transport")
merged1
```

```python
b= b.set_index("Transport")
b
```

```python
merged2 = pd.merge(a,b , left_index = True , right_index=True)
merged2
```

```python
merged3 = pd.merge(a,b , on="Transport" , how='outer')
merged3
```
